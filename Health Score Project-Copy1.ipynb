```python
import pandas as pd

#Reading raw data
data = pd.read_csv('newest_csv.csv')
display(data)
```

```python
#Extracting only EUIs and SIBs
data =  data.loc[data['EUI_OR_SIB'].isin(['EUI','SIB'])]
```

```python
data_sus=data[data['HOUR_TO_COMPLETION'].isnull()]

#Extracting out NaNs in 'Hour To Completion'
data_sus.to_csv('nans.csv', index=False)
```

```python
#Extracting relevant variables
data_analy = data[["ISSUE_KEY","PLATFORM","PRODUCT","RESOLVED_TIMESTAMP","SEVERITY", "GROUP_NAMES_SEPARATED","COE_NAMES_SEPARATED_COPY","RESOLUTION_TIME","HOUR_TO_COMPLETION"]]
```

```python
data_analy['HOUR_TO_COMPLETION'] = data_analy['HOUR_TO_COMPLETION'].replace(0,1) 
#Replacing zero with one so to avoid inf values
```

```python
#Removing NA Values
data_analy = data_analy[data_analy['COE_NAMES_SEPARATED_COPY'].notna()]
data_analy = data_analy[data_analy['GROUP_NAMES_SEPARATED'].notna()]
data_analy = data_analy[data_analy['RESOLUTION_TIME'].notna()]
data_analy = data_analy[data_analy['HOUR_TO_COMPLETION'].notna()]
```

```python
#Using a loop to calculate each COE's variables by SEVERITY

import statistics as st

#Empty Variables for later
COE_EUIs=[]
odds=[]
dev=[]
meds=[]
k=0


COE_names=data_analy['COE_NAMES_SEPARATED_COPY'].unique() #Getting all COE names
sev_names=data_analy['SEVERITY'].unique() #Getting all different severity levels.
sev_names.sort()

COE_tot_sev2=pd.DataFrame(COE_names,columns=['COE_names']) #Making DFs in advance for each Severity level
COE_tot_sev3=pd.DataFrame(COE_names,columns=['COE_names'])
COE_tot_sev4=pd.DataFrame(COE_names,columns=['COE_names'])


for s in sev_names:
    if s=='P1': #Not all COEs have P1, therefore it has is own code.
        x=data_analy.loc[data_analy['SEVERITY'] == s]
        COE_names1=x['COE_NAMES_SEPARATED_COPY'].unique()
        COE_tot_sev1=pd.DataFrame(COE_names1,columns=['COE_names'])
        for i in COE_names1:
            p=x.loc[x['COE_NAMES_SEPARATED_COPY'] == i]
            COE_EUIs.append(len(p))
            for j in range(0,len(p)):
                odd=(p.iloc[j,8])/(p.iloc[j,7])
                odds.append(odd)
                if p.iloc[j,8] > p.iloc[j,7]:
                    k=k+1
            meds.append((k/len(p))*100)
            dev.append(st.median(odds))
            k=0
            odds=[]
        COE_tot_sev1["P1-EUI Num"]=COE_EUIs
        COE_tot_sev1["P1-Breach%"]=meds
        COE_tot_sev1["P1-Deviation"]=dev
        COE_EUIs=[]
        meds=[]
        dev=[]
    if s=='P2':
        x=data_analy.loc[data_analy['SEVERITY'] == s]
        for i in COE_names:
            p=x.loc[x['COE_NAMES_SEPARATED_COPY'] == i]
            COE_EUIs.append(len(p))
            for j in range(0,len(p)):
                odd=(p.iloc[j,8])/(p.iloc[j,7])
                odds.append(odd)
                if p.iloc[j,8] > p.iloc[j,7]:
                    k=k+1
            meds.append((k/len(p))*100)
            dev.append(st.median(odds))
            k=0
            odds=[]
        COE_tot_sev2["P2-EUI Num"]=COE_EUIs
        COE_tot_sev2["P2-Breach%"]=meds
        COE_tot_sev2["P2-Deviation"]=dev
        COE_EUIs=[]
        meds=[]
        dev=[]
    if s=='P3':
        x=data_analy.loc[data_analy['SEVERITY'] == s]
        for i in COE_names:
            p=x.loc[data_analy['COE_NAMES_SEPARATED_COPY'] == i]
            COE_EUIs.append(len(p))
            for j in range(0,len(p)):
                odd=(p.iloc[j,8])/(p.iloc[j,7])
                odds.append(odd)
                if p.iloc[j,8] > p.iloc[j,7]:
                    k=k+1
            meds.append((k/len(p))*100)
            dev.append(st.median(odds))
            k=0
            odds=[]
        COE_tot_sev3["P3-EUI Num"]=COE_EUIs
        COE_tot_sev3["P3-Breach%"]=meds
        COE_tot_sev3["P3-Deviation"]=dev
        COE_EUIs=[]
        meds=[]
        dev=[]
    if s=='P4':
        x=data_analy.loc[data_analy['SEVERITY'] == s]
        for i in COE_names:
            p=x.loc[x['COE_NAMES_SEPARATED_COPY'] == i]
            COE_EUIs.append(len(p))
            for j in range(0,len(p)):
                odd=(p.iloc[j,8])/(p.iloc[j,7])
                odds.append(odd)
                if p.iloc[j,8] > p.iloc[j,7]:
                    k=k+1
            meds.append((k/len(p))*100)
            dev.append(st.median(odds))
            k=0
            odds=[]
        COE_tot_sev4["P4-EUI Num"]=COE_EUIs
        COE_tot_sev4["P4-Breach%"]=meds
        COE_tot_sev4["P4-Deviation"]=dev
```
